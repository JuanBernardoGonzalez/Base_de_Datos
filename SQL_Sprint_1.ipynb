# Ejercicio 1

El esquema describe una base de datos de películas con cinco tablas relacionadas entre ellas. 
### Tablas
1. tb_genre: tiene cinco variables y describe principalmente el género de las películas. genre_id es la *llave primaria* y acepta *integers* como datos. El resto de datos que acepta son 2 de caractéres (varing 40 y 10) y 2 más de fechas.
2. tb_movie: tiene ocho variales y describe principalmente las películas (título, fecha de la película y su formato. movie_id es la *llave primaria* y acepta *integers* como datos. movie_genre_id es la *llave extrangera* que se relaciona con la variable genre_id de la tabla tb_movie. El resto de datos que acepta son 3 de fechas y 3 de caractéres (varing 100, 50 y 10). 
3. tb_role: tiene 5 variables y describe principalmente el papel que desempeñan las persona. role_id es la *llave primaria*, acepta *integers* como datos y se relaciona con la tabla tb_movie_person. El resto de datos que acepta son 2 de caractéres (varyng 60 y 10) y dos de fechas. 
4. tb_person: tiene nueve variables y describe principalmente el nombre de las personas, nacionalidad, fecha de nacimiento y fecha de defunción. person_id es la *llave primaria* y acepta *integers* como datos. person_parent_id es la *llave extrangera* que se relaciona con la variable person_id de la misma tabla. El resto de datos son 4 de fechas y 3 de caractéres (varing 100, 40 y 10). 
5. tb_movie_person: tiene 7 variables y describe ls identificación de la película, la persona y su papel, además de informarnos si la película ha tenido algún premio.Tiene tres *llaves primarias*: movie_id, person_id y role_id, todas aceptan *integers* como datos. Estas se relacionan con las *llaves* de las tablas tb_movie, tb_person y tb_role, respectivamente. El resto de datos que acepta son 2 de caractéres y 2 de fechas.
